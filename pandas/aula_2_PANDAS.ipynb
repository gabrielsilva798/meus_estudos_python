## como apagar, excluir linhas de acordo com uma determinada condição.
    - pra excluir ou selecionar somente linhas com uma certa condição é só colocar no final do comando loc. a condição. 
    - se vc quer EXCLUIR números menores que 160 é só pegar os MAIORES que 160. USE ESSA LÓGICA PARA EXCLUIR LINHAS.
    - se vc que linhas !=, ==, <= ou >= da condição desejada, aqui é que fossem maior de 160.

```python
valores160 = vendas_df.loc[vendas_df['Valor Final']>=160]
valores160.head(10)
```

```python
#agora além de pegar todas as LINHAS com Norte Shopping eu vou pegar também as COLUNAS Data, Produto e Quantidade.
#que tenha Norte Shopping
#pegar várias linhas e colunas usando o .loc
coluna_e_linhas = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ['Data', 'Produto', 'Quantidade']]
coluna_e_linhas.head(10)
# o padrãodo loc é .loc[linhas, colunas] ,mas se vc só colocar um ele é automaticamente lido como linha.
```

```python
#como dito para achar algo específico na tabela com o loc.(DataFrame[linha, coluna])
#voce coloca o DataFrame e dentro dos colchetes a linha e coluna do que vpce quer.
display(vendas_df.loc[1, 'Produto'])
print(vendas_df.loc[1, 'Produto'])
# com display tem aspase com print não tem.
```

# o método DataFrame.sort_values coloca uma planilha em ordem do menor para o maior.

## o group by serve para fazer a comparação de colunas.

### o group by agrupa os valores iguais de uma coluna e soma outra coluna relacioanda a elas no caso ele agrupa os produtos da coluna PRODUTO e soma a coluna VALOR FINAL ligadas a elas para conseguir o faturamento veja abaixo.

```python
#eu vou agrupar a coluna Produto, e as colunas númericas fora dos parentes serão somadas pelo .sum()
faturamento_produto = vendas_df.groupby('Produto').sum()
#coloquei na variável faturamento_produto
faturamento_produto.head()
```

## para pegar só colunas específicas

```python
#ele vai pegar somente a coluna PRODUTOS e VALOR FINAL.
faturamento_produto_final = vendas_df[['Produto','Valor Final']].groupby('Produto').sum()
display(faturamento_produto_final)
#ele podia mostrar a soma, a média....
```

## ele calcula a média da coluna QUANTIDADE para cada produto da coluna PRODUTO.


```python
media = vendas_df.groupby('Produto')['Quantidade'].median()
media.head(20)
```

## mesclar 2 DataFrame

```python
#vou importar a tabela gerente agora
gerentes_df = pd.read_excel('Gerentes.xlsx')
display(gerentes_df)
```

```python
#eu quero adicionar na tabela original uma tabela de gerente
#automaticamente ele já ver as colunas iguais (o ID LOja), e usa como parâmetro para ligar os gerentes as suas lojas específicas
vendas_df = vendas_df.merge(gerentes_df)
vendas_df.head()
```
